```python
import pandas as pd
import matplotlib.pyplot as plt
import glob 

def plot_convergence(scheme_name, k, T, N_list):    
    plt.style.use('seaborn-v0_8-whitegrid')
    fig, ax = plt.subplots(figsize=(12, 7))

    colors = plt.cm.viridis_r([i / len(N_list) for i in range(len(N_list))])

    for i, n_val in enumerate(N_list):
        filename = f"../HW2/{scheme_name}_k{k}_T{T}_N{n_val}.csv"
        data = pd.read_csv(filename)
        # 
        ax.plot(data['x'], data['numerical_u'], 'o-', 
                label=f'Numerical (N={n_val})', 
                color=colors[i], 
                markersize=3, 
                linewidth=1.5,
                alpha=0.8) 
        
            
    
    highest_n = max(N_list)
    filename_exact = f"../HW2/{scheme_name}_k{k}_T{T}_N{highest_n}.csv"
    data_exact = pd.read_csv(filename_exact)
    ax.plot(data_exact['x'], data_exact['exact_u'], '--', 
                label='Expectation', 
                color='red', 
                linewidth=2.5,
                zorder=10)
    
    ax.set_xlabel('x', fontsize=12)
    ax.set_ylabel('u(x, t)', fontsize=12)
    ax.legend(fontsize=10)
    ax.grid(True, which='both', linestyle='--', linewidth=0.5)

    output_image_filename = f"plot_{scheme_name}_k{k}_T{T}.png"
    plt.tight_layout()
    plt.savefig(output_image_filename, dpi=300)
    #plt.savefig("test.png", dpi=300)
    #print(f"plot saved to '{output_image_filename}'")
    plt.show()


if __name__ == "__main__":
    # plot_convergence(
    #     scheme_name='scheme2',
    #     k=10,
    #     T=10,
    #     N_list=[1280, 2560, 5120,10240]
    # )

    # plot_convergence(
    #     scheme_name='scheme2',
    #     k=1,
    #     T=10,
    #     N_list=[20, 40, 80, 160]
    # )

    # plot_convergence(
    #     scheme_name='scheme2',
    #     k=5,
    #     T=10,
    #     N_list=[20, 40, 80, 160]
    # )

    # plot_convergence(
    #     scheme_name='scheme2',
    #     k=10,
    #     T=10,
    #     N_list=[20, 40, 80, 160]
    # )

    plot_convergence(
        scheme_name='scheme3',
        k=1,
        T=5,
        N_list=[640]
    )
```
